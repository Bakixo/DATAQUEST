```python
import pandas as pd

series = pd.read_csv("WDISeries.csv", low_memory=False)

keep_codes = [
    "SP.DYN.LE00.IN",      # Life expectancy
    "NY.GDP.PCAP.KD",      # GDP per capita, constant
    "SE.SEC.ENRR",         # Secondary enrollment
    "EN.ATM.CO2E.PC",      # CO2 per capita
    "EG.USE.PCAP.KG.OE",   # Energy use per capita
    "EN.ATM.PM25.MC.M3"    # PM2.5 exposure
]

selected = series[series["Series Code"].isin(keep_codes)][[
    "Indicator Name",
    "Series Code",
    "Topic",
    "Unit of measure",
    "Short definition"
]]

print(selected)
```
